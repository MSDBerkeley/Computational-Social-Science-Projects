## 2. Data Preprocessing and Cleaning

```python
# Check the column names in the dataset
print(chicago_inspections_2011_to_2013.columns)
```

```python
# drop datetime info
# -----------

chicago_inspections_2011_to_2013 = chicago_inspections_2011_to_2013.dropna().drop(['Inspection_Date',
                                                                                   'minDate',
                                                                                   'maxDate',
                                                                                   'Inspection_MonthYear'],
                                                                                  axis = 1)
```

```python
# process target 
# -----------
y = chicago_inspections_2011_to_2013['Results']

# decide if you want to binarize the outcome variable 
# -----------
# comment out the following lines of code if you don't want to binarize the target variable
y = y.replace({'Pass w/ Conditions': 'Pass'})
lb_style = LabelBinarizer()
y = lb_style.fit_transform(y)

# recode 0s and 1s so 1s are "Fail"
y = np.where(y == 1, 0 ,1)


# process features
# -----------

# create feature dataset
X = chicago_inspections_2011_to_2013.drop(columns = ['Results', 
                                                     'Facility_Type', 
                                                     'LICENSE_DESCRIPTION', 
                                                     'WARD_PRECINCT'])


# get dummies
X = pd.get_dummies(X)
```

### Model 1: Logistic Regression

Logistic regression is a linear model that predicts the probability of a binary outcome using a logistic function. It assumes a linear relationship between features and the log-odds of the outcome.

- Pros: Simple to implement, interpretable coefficients, and computationally efficient.
- Cons: Limited to linear decision boundaries; performance may degrade with complex, non-linear relationships.

Suitability: Good baseline model for classification tasks with binary labels.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

# Logistic Regression hyperparameter tuning
param_grid_log_reg = {'C': [0.01, 0.1, 1, 10, 100]}
log_reg = LogisticRegression(max_iter=1000, random_state=42)
log_reg_cv = GridSearchCV(log_reg, param_grid_log_reg, cv=5, scoring='f1')
log_reg_cv.fit(X, y)

# Define the best Logistic Regression model
best_log_reg = log_reg_cv.best_estimator_

# Best hyperparameters and F1 score
print("Best Hyperparameters (Logistic Regression):", log_reg_cv.best_params_)
print("Best Mean F1 Score (Logistic Regression):", log_reg_cv.best_score_)
```

###  Model 2: Random Forest

Random Forest is an ensemble method that builds multiple decision trees and combines their outputs to improve predictive performance. It assumes that a collection of uncorrelated trees will produce better results.

- Pros: Handles non-linear relationships, robust to overfitting, and works well with high-dimensional data.
- Cons: Less interpretable than simpler models and computationally intensive for large datasets.

Suitability: Effective for capturing complex interactions and non-linear relationships in features.

```python
from sklearn.ensemble import RandomForestClassifier

# Random Forest hyperparameter tuning
param_grid_rf = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10]
}
rf = RandomForestClassifier(random_state=42)
rf_cv = GridSearchCV(rf, param_grid_rf, cv=5, scoring='f1')
rf_cv.fit(X, y)

# Define the best Random Forest model
best_rf = rf_cv.best_estimator_

# Best hyperparameters and F1 score
print("Best Hyperparameters (Random Forest):", rf_cv.best_params_)
print("Best Mean F1 Score (Random Forest):", rf_cv.best_score_)
```

### Model 3: AdaBoost

AdaBoost combines multiple weak classifiers (e.g., decision stumps) by iteratively re-weighting instances to focus on those misclassified by previous classifiers.

- Pros: Handles non-linear relationships, adaptive to difficult instances, and works well with smaller datasets.
- Cons: Prone to overfitting if not tuned properly and can be sensitive to noise.

Suitability: Effective for improving predictive performance on imbalanced datasets or with weak classifiers.

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import GridSearchCV

# AdaBoost hyperparameter tuning
param_grid_ada = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 1]
}

# Initialize AdaBoost with SAMME algorithm to correct SAMME.R discontinuation error
ada = AdaBoostClassifier(algorithm='SAMME', random_state=42)

# Perform GridSearchCV with 5-fold cross-validation
ada_cv = GridSearchCV(ada, param_grid_ada, cv=5, scoring='f1')
ada_cv.fit(X, y)

# Define the best AdaBoost model
best_ada = ada_cv.best_estimator_

# Print the best hyperparameters and mean F1 score
print("Best Hyperparameters (AdaBoost):", ada_cv.best_params_)
print("Best Mean F1 Score (AdaBoost):", ada_cv.best_score_)
```

### Validation Metrics

Be sure to explain which of these metrics you would want to prioritize when conducting predictive auditing in this context and why.

**Hint**: Try writing a for loop to use [`cross_val_score()`](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.cross_val_score.html) to check for accuracy, precision, recall and f1 across all of your models.

```python
from sklearn.model_selection import cross_val_score

# Define a function to evaluate validation metrics
def evaluate_model_metrics(model, X, y, cv=5):
    metrics = ['accuracy', 'precision', 'recall', 'f1']
    results = {}
    
    for metric in metrics:
        # Perform cross-validation for each metric
        scores = cross_val_score(model, X, y, cv=cv, scoring=metric)
        results[metric] = {
            'mean': scores.mean(),
            'std': scores.std()
        }
    
    return results
```

```python
# Evaluate Logistic Regression
print("Logistic Regression Metrics:")
log_reg_metrics = evaluate_model_metrics(best_log_reg, X, y)
for metric, result in log_reg_metrics.items():
    print(f"{metric.capitalize()} - Mean: {result['mean']:.4f}, Std: {result['std']:.4f}")

# Evaluate Random Forest
print("\nRandom Forest Metrics:")
rf_metrics = evaluate_model_metrics(best_rf, X, y)
for metric, result in rf_metrics.items():
    print(f"{metric.capitalize()} - Mean: {result['mean']:.4f}, Std: {result['std']:.4f}")

# Evaluate AdaBoost
print("\nAdaBoost Metrics:")
ada_metrics = evaluate_model_metrics(best_ada, X, y)
for metric, result in ada_metrics.items():
    print(f"{metric.capitalize()} - Mean: {result['mean']:.4f}, Std: {result['std']:.4f}")
```

### Analysis

The models performed well overall, with accuracies above 92%, but accuracy alone is not the best metric for predictive auditing. Instead, precision, recall, and F1 score provide deeper insights:

1. **Precision** measures how often predicted "Failures" are correct. While Random Forest had relatively strong precision (76.92%), AdaBoost's lower precision (73.60%) indicates a slightly higher rate of unnecessary audits.
2. **Recall** is crucial in predictive auditing to minimize false negatives, ensuring high-risk establishments are identified. AdaBoost achieved the highest recall (95.66%), making it the best choice for prioritizing audits effectively.
3. **F1 Score** balances precision and recall. AdaBoost (83.16%) and Random Forest (83.00%) performed similarly, with AdaBoost slightly ahead.

#### Which Metric to Prioritize and Why? 

For predictive auditing, recall should be prioritized. Missing high-risk establishments (false negatives) poses significant risks, making recall the most critical metric. AdaBoost, with its recall of 95.66%, is the best model for this context.

If balancing false positives and false negatives is also important, F1 Score can guide the decision. AdaBoost‚Äôs highest F1 score (83.16%) reinforces its suitability.

## 5. Discussion Questions

1. Why do we need metrics beyond accuracy when using machine learning in the social sciences and public policy?


Accuracy alone does not provide a full picture of a model's performance, particularly in contexts like social sciences and public policy, where the consequences of incorrect predictions can vary significantly. For instance, in imbalanced datasets (e.g., where most inspections Pass and few Fail), a high accuracy might simply reflect the model's ability to predict the majority class while failing to detect critical minority outcomes. Metrics like precision and recall help assess the trade-offs between false positives and false negatives, which is vital in decision-making processes. For example, a high recall ensures failures are not overlooked, while precision minimizes wasted resources on unnecessary audits. Additionally, the F1 score provides a balanced view by combining precision and recall, which is especially useful when false positives and false negatives have significant policy implications.

2. Imagine that establishments learned about the algorithm being used to determine who gets audited and they started adjusting their behavior (and changing certain key features about themselves that were important for the prediction) to avoid detection. How could policymakers address this interplay between algorithmic decisionmaking and real world behavior?

When establishments adjust their behavior to avoid detection, it creates a dynamic environment where the effectiveness of static algorithms diminishes over time. Policymakers can address this by incorporating adaptive strategies into algorithmic systems. First, algorithms should be periodically retrained using updated data to account for evolving behavior patterns. Second, policymakers could use ensemble methods or randomized feature weighting to make the decision criteria less predictable. Additionally, combining algorithmic predictions with human oversight ensures nuanced decision-making that algorithms alone cannot provide. Finally, policymakers should consider the root causes of such evasive behavior and address them through systemic reforms, such as stricter penalties for manipulation or incentives for compliance. This interplay underscores the importance of continually monitoring and refining machine learning systems to maintain their integrity and relevance in public policy applications. Finally -and less popular- introducing "black box" elements to the algorithm (where not all decision rules are transparent) can prevent establishments from reverse-engineering the system.
