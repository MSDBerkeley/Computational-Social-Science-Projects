### Train PCA and Discuss Number of Components

```python
from sklearn.decomposition import PCA

# Train PCA on the current nhanes_scaled (features only)
pca = PCA()
pca.fit(nhanes_scaled)

# Explained variance ratio
explained_variance = pca.explained_variance_ratio_
print(f"Explained Variance Ratio: {explained_variance}")
```

### Barplot of Components

```python
# Barplot of explained variance by components
plt.figure(figsize=(10, 6))
plt.bar(range(1, len(explained_variance) + 1), explained_variance)
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
plt.title('Variance Explained by Principal Components')
plt.show()
```

```python
# Cumulative explained variance
cumulative_variance = np.cumsum(pca.explained_variance_ratio_)

# Plot cumulative explained variance
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Variance Explained')
plt.title('Cumulative Explained Variance by Principal Components')
plt.axhline(y=0.9, color='r', linestyle='--', label='90% Variance Threshold')
plt.legend()
plt.show()
```

```python
# Find the number of components to explain 90% variance
num_components_90 = np.argmax(cumulative_variance >= 0.9) + 1
print(f"Number of components to explain 90% variance: {num_components_90}")
```

### How many components will you use to fit a supervised learning model?

```python
# Fit PCA with 120 components
n_components = 120  # Based on cumulative variance analysis
pca = PCA(n_components=n_components)
X_pca = pca.fit_transform(nhanes_scaled)

print(f"Shape of PCA-transformed data: {X_pca.shape}")
```

**_Choose the number of components and provide 1-2 sentences about your choice of the number of components._**

I chose 120 components because they collectively explain 90% of the variance in the dataset. This threshold ensures that most of the meaningful information in the data is retained while significantly reducing the dimensionality from the original feature set.

### 2D Scatterplot of the first two components

```python
# Scatterplot for the first two components
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], alpha=0.5)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('2D Scatterplot of the First Two Principal Components')
plt.show()
```

**Analysis of the 2D Scatterplot of the First Two Principal Components**

The scatterplot of the first two principal components shows that most data points are concentrated in a dense cluster near the origin, with a few points spread more widely. This indicates that while the first two components capture some variance, they may not fully differentiate meaningful patterns in the data. ### Choose a Clustering Algorithm

I chose the K-Means clustering algorithm because it is efficient and works well for partitioning data into distinct groups based on similarity. It minimizes the within-cluster sum of squares (WCSS) by assigning points to the nearest cluster centroid.

### Cluster nhanes

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Determine the optimal number of clusters using the elbow method
wcss = []
for k in range(1, 11):  # Test k from 1 to 10
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(nhanes_scaled)
    wcss.append(kmeans.inertia_)

# Plot the elbow method results
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.title('Elbow Method for Optimal k')
plt.show()
```

**Choice of k:**

The elbow point appears to occur at k = 7, where the Within-Cluster Sum of Squares (WCSS) starts to flatten. Beyond this point, the reduction in WCSS diminishes, indicating that 7 clusters provide a good balance between minimizing WCSS and avoiding overfitting.

```python
# Train K-Means with k = 7
k = 7  # Chosen number of clusters
kmeans = KMeans(n_clusters=k, random_state=42)
clusters = kmeans.fit_predict(nhanes_scaled)

# Add the cluster labels to the dataset
nhanes_scaled['Cluster'] = clusters

# Check the first few rows with clusters
print(nhanes_scaled.head())
```

### Neural Network on Original Features

```python
# partition data
# -----------
#y = ... # either hsd010 or hsd010_binary, may need to convert to numeric if it isn't already 
#X = ... # drop out any columns that aren't features

import pandas as pd
from sklearn.model_selection import train_test_split

# Reload the original dataset
nhanes_original = pd.read_csv('nhanes.csv')

# Extract the target variable HSD010_binary
nhanes_original['HSD010_binary'] = nhanes_original['HSD010'].replace(
    [1, 2, 3, 4, 5],  # Original health ratings
    [1, 1, 1, 0, 0]   # Binary mapping: 1 (good), 0 (poor)
)

# Set target variable (y)
y = nhanes_original['HSD010_binary']

# Set feature variables (X): Drop columns that are not predictors
X = nhanes_original.drop(columns=['SEQN', 'HSD010', 'HSD010_binary'], errors='ignore')

# Partition the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    test_size=0.25, 
                                                    random_state=10)

# Check the shapes of the training and testing sets
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")
```

```python
# load libraries
# -----------
import keras
from keras.utils import to_categorical

num_classes = len(y_train.unique())
# convert class vectors to binary class matrices
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)
num_classes = y_test.shape[1]

# Print the shape of the one-hot encoded targets to verify
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")
print(f"Number of classes: {num_classes}")
```

```python
from keras.layers import Input
from keras.models import Sequential
from keras.layers import Dense, Dropout

# create neural network model
# -----------

# Initialize the model
model = Sequential()

# Add Input layer explicitly
model.add(Input(shape=(X_train.shape[1],)))

# Add hidden layers
model.add(Dense(64, kernel_initializer='he_normal', activation='relu'))
model.add(Dense(32, kernel_initializer='he_normal', activation='relu'))

# Add output layer for classification
model.add(Dense(num_classes, kernel_initializer='he_normal', activation='softmax'))

# Compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
history = model.fit(X_train, y_train, 
                    validation_data=(X_test, y_test), 
                    epochs=20, batch_size=32, verbose=1)
```

## Discussion Questions

1. In your own words, what is the difference between PCA and clustering?

PCA (Principal Component Analysis) is a method that reduces the number of variables in a dataset by focusing on the most important features, helping to simplify the data. It helps to remove unnecessary details and keeps the essential patterns. On the other hand, clustering groups data into categories based on their similarities. Instead of simplifying the data, clustering organizes it into meaningful groups without needing any labels or pre-defined categories.

Yes, combining PCA and clustering had advantages. PCA helped simplify the data by reducing the number of features, making it easier to analyze. Adding cluster membership as a new feature allowed the model to identify groups in the data, which gave it more context. This helped improve the model’s ability to generalize and perform well on new data. However, the improvements were modest, likely because the data still had some complexity that the combination of PCA and clustering couldn't fully address.

3. How can unsupervised techniques help with downstream supervised learning tasks when working with "big data?"

Unsupervised techniques like PCA and clustering are very helpful when working with large datasets. PCA makes the data simpler by focusing on the most important information, which speeds up analysis and makes it easier to understand. Clustering helps by grouping similar data points together, making it easier to spot trends or patterns. When used together, these techniques help to reduce the complexity of big data, making it more manageable and improving the accuracy of predictions for decision-making.
